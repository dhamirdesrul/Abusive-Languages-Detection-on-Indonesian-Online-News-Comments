```python
import pandas as pd
```

```python
import urllib.request
import json

save_comment = []
for i in range(9):
    url = 'https://newcomment.detik.com/graphql?query={%20search(type:%20%22comment%22,size:%2010%20,page:'+(str(i+1))+',sort:%22newest%22,%20adsLabelKanal:%20%22detik_news%22,%20adsEnv:%20%22desktop%22,%20query:%20[{name:%20%22news.artikel%22,%20terms:%20%224707640%22%20}%20,%20{name:%20%22news.site%22,%20terms:%20%22dtk%22}%20])%20{%20paging%20sorting%20counter%20counterparent%20profile%20hits%20{%20posisi%20hasAds%20results%20{%20id%20author%20content%20like%20prokontra%20status%20news%20create_date%20pilihanredaksi%20refer%20liker%20{%20id%20}%20reporter%20{%20id%20status_report%20}%20child%20{%20id%20child%20parent%20author%20content%20like%20prokontra%20status%20create_date%20pilihanredaksi%20refer%20liker%20{%20id%20}%20reporter%20{%20id%20status_report%20}%20authorRefer%20%20}%20%20}%20%20}%20%20}%20%20}'
    contents = urllib.request.urlopen(url).read()
    diction = json.loads(contents)
    for i in diction['data']['search']['hits']['results']:
        
        save_comment.append(i['content'])
    # save_comment
```

```python
save_comment
```

```python
save_comment = pd.DataFrame(save_comment, index=None)
```

```python
save_comment.to_csv('Menolak Lupa Janji Jokowi soal Pemberantasan Korupsi.csv')
```
