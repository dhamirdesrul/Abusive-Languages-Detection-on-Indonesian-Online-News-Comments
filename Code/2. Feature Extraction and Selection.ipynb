```python
def extract_words(sentence):
    ignore_words = ['or']
    words = re.sub("[^\w]", " ",  sentence).split() #nltk.word_tokenize(sentence)
    words_cleaned = [w.lower() for w in words if w not in ignore_words]
#     print(words_cleaned)
    return words_cleaned    

def tokenize_sentences(sentences):
    words = []
    for sentence in sentences:
        w = extract_words(sentence)
        words.extend(w)
        
    words = sorted(list(set(words)))
    return words

def bagofwords(sentence, words):
    sentence_words = extract_words(sentence)
    # frequency word count
    bag = np.zeros(len(words))
    for sw in sentence_words:
        for i,word in enumerate(words):
            if word == sw: 
                bag[i] += 1
                
    return np.array(bag)
```

```python
def random_match(set_source):
    atribut, label = split_count(set_source)
    random_score = []
    len_prob = len(atribut) * len(label)
    for j in atribut:
#         print(j)
        for k in label:
#             print(k)
            temp = [j, k]
            random_score.append(temp)
    return random_score
```

```python
def split_count(set_source):
    atribut, label = [],[]
    for i in set_source:
        atribut.append(i[0])
        label.append(i[1])
    count_atribut = Counter(atribut)
    count_label = Counter(label)
    return count_atribut, count_label
```

```python
def split_no_array(label):
    no_label = []
    for i in label:
        no_label.append(i[0])
    return no_label
```

```python
def random_probability_score(x_y, set_source, label_fix, var_fix):
    value = random_match(set_source)
    atribut, label = split_count(set_source)
    tot_document = len(x_y)
    freq = {}
    for i in value:
    #     print(i)
        for j in x_y:
    #         print(j)
            if ((i[0] == j[0]) and (i[1] == j[1])):
                if (i[0], i[1]) in freq:
                    freq[(i[0], i[1])] += 1
                else:
                    freq[(i[0], j[1])] = 1
    prob_x_y, tot_MI = [], []
    for key, value in freq.items():
        xy = value/tot_document
        prob_x_y.append([key, xy])
        for k_var, v_var in var_fix.items():
            if k_var == key[0]:
                temp_var = v_var
        for k_label, v_label in label_fix.items():
            if k_label == key[1]:
                temp_label = v_label
        px = temp_var/tot_document
        py = temp_label/tot_document
        t_px_py = px*py
        a = math.log2(xy/(t_px_py))
        tot_MI.append([key, a])
    return tot_MI
```

```python
def random_probability_score_new(x_y, set_source, label_fix, var_fix):
    value = random_match(set_source)
    tot_document = len(x_y)
    freq = {}
    for i in value:
        for j in x_y:
            if ((i[0] == j[0]) and (i[1] == j[1])):
                if (i[0], i[1]) in freq:
                    freq[(i[0], i[1])] += 1
                else:
                    freq[(i[0], j[1])] = 1
    prob_x_y, tot_MI = [], []
    for key, value in freq.items():
        xy = value/tot_document
        prob_x_y.append([key, xy])
        a = 0
        for k_var, v_var in var_fix.items():
            if k_var == key[0]:
                temp_var = v_var
        for k_label, v_label in label_fix.items():
            if k_label == key[1]:
                temp_label = v_label
        px = temp_var/tot_document
        py = temp_label/tot_document
        t_px_py = px*py
        a = math.log(xy/(t_px_py))
        tot_MI.append([key, a])
    return tot_MI
```

```python
def create_data_source(var, label):
    x_y = []
    set_source = []
    for i in range(len(var)):
        x_y.append([var[i][0],label[i][0]])
    # x_y = list(x_y)
    for i in x_y:
        if i not in set_source:
            set_source.append(i)
    return x_y, set_source
```

## Melakukan Mutual Information

```python
def tot_mi_array(tot, df_ekstraksifitur):
    list_hasil = []
    maks = int(max(df_ekstraksifitur.max(axis=1)))
    for i in tot:
        arr = np.zeros((maks+1,4))
        for j in i:
            label = int(j[0][1])
            atr = int(j[0][0])
            value = j[1]
#             print(atr, label, value)
            arr[atr][label] = value
        list_hasil.append(arr)
    return list_hasil
```
