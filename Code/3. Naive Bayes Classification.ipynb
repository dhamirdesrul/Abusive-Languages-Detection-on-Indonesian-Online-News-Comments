```python
import pandas as pd
import numpy as np
import collections
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn import metrics
import csv
from sklearn.model_selection import RepeatedKFold
```

```python
# Python script for confusion matrix creation. 
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

from sklearn.metrics import f1_score

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
```

```python
load_file = pd.read_csv('bagofwords_no stem&nosw.csv', index_col= None)
load_file = load_file.drop('Unnamed: 0', axis=1)
```

```python
load_file
```

```python
len(load_file)
```

```python
value_ekstraksi = load_file.drop('Label', axis=1)
# value_ekstraksi
```

```python
label_ekstraksi = load_file['Label']
# label_ekstraksi
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(value_ekstraksi, label_ekstraksi, test_size=0.33) # 70% training and 30% test
```

```python
# y_train
```

```python
gnb = GaussianNB()

gnb.fit(X_train, y_train)

ab = gnb.predict(X_test)

y_pred = ab
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
y_true = y_test
print(f1_score(y_true, y_pred, average='macro'))
print(f1_score(y_true, y_pred, average='micro'))
print(f1_score(y_true, y_pred, average='weighted')) 
print(f1_score(y_true, y_pred, average=None))
```

```python
results = confusion_matrix(y_true, y_pred) 
print('Confusion Matrix :')
print(results) 
print('Accuracy Score :',accuracy_score(y_true, y_pred))
print('Report : ')
print(classification_report(y_true, y_pred))
```
