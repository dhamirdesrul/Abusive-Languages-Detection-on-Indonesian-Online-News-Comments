```python
value_ekstraksi = load_file_wsw.drop('Label', axis=1)
# value_ekstraksi
```

```python
label_ekstraksi = load_file_wsw['Label']
# label_ekstraksi
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(value_ekstraksi, label_ekstraksi, test_size=0.3) # 70% training and 30% test
```

```python
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=4)

#Train the model using the training sets
knn.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
y_true = y_test
print(f1_score(y_true, y_pred, average='macro'))
print(f1_score(y_true, y_pred, average='micro'))
print(f1_score(y_true, y_pred, average='weighted')) 
print(f1_score(y_true, y_pred, average=None))
```

```python
results = confusion_matrix(y_true, y_pred) 
print('Confusion Matrix :')
print(results) 
print('Accuracy Score :',accuracy_score(y_true, y_pred))
print('Report : ')
print(classification_report(y_true, y_pred))
```
